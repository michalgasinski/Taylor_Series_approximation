```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return x**2 / np.exp(x)
x0=2.5
h_seq = np.arange(0,10,0.01)
```

```python
def der(x,h=10**(-6)):
     return (fun(x+h)-fun(x))/h
def der2(x,h=10**(-6)):
     return (fun(x+2*h)-2*fun(x+h)+fun(x))/h**2
```

```python
def taylor1(x,h):
    return fun(x) + der(x)*(h-x)
def taylor2(x,h):
    return fun(x) + der(x)*(h-x) + 0.5*der2(x)*((h-x))**2
```

```python
plt.plot(h_seq, fun(h_seq), label="f(x)", color='black')
plt.plot(h_seq, taylor1(x0, h_seq), label="Taylor 1", color='red')
plt.plot(h_seq, taylor2(x0, h_seq), label="Taylor 2", color='blue')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')  # Customize legend location
plt.ylim([0, 0.8])

# Show the plot
plt.show()
```
